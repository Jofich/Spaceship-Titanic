```python
def split_cabin(cabin):
    
    cabin = cabin.split('/')
    
    return cabin[0],int(cabin[1]),cabin[2]
```

```python
deck = []
num = []
side = []

for x in train['Cabin']:
    
    temp_deck, temp_num, temp_side = split_cabin(x)
    deck.append(temp_deck)
    num.append(temp_num)
    side.append(temp_side)
```

```python
train['Deck'] = deck
train['Num'] = num
train['Side'] = side
```

```python
deck = []
num = []
side = []

for x in test['Cabin']:
    
    temp_deck, temp_num, temp_side = split_cabin(x)
    deck.append(temp_deck)
    num.append(temp_num)
    side.append(temp_side)
```

```python
test['Deck'] = deck
test['Num'] = num
test['Side'] = side
```

```python
columns = ['Name','Cabin','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','PassengerId']
test = test.drop(columns = columns)
train = train.drop(columns = columns)
```

```python
train.head()
```

```python
categ = train.loc[:, train.dtypes == object].columns
for i in categ:
    print(i)
    print(set(test[i].unique()).symmetric_difference(set(train[i].unique())))
```

```python
from sklearn import preprocessing

categ = train.loc[:, train.dtypes == object].columns

for i in categ:
    le = preprocessing.LabelEncoder()

    le.fit(train[i])

    train[i] = le.transform(train[i])
    test[i] = le.transform(test[i])
```

```python
le = preprocessing.LabelEncoder()
le.fit(train['Transported'])
train['Transported'] = le.transform(train['Transported'])
```

```python
X = train.drop(columns=['Transported'])
y = train['Transported']
```

```python
X.shape, y.shape
```

```python
from imblearn.over_sampling import SMOTE

os = SMOTE(random_state = 123, k_neighbors = 4)

column = X.columns

X_os, y_os = os.fit_resample(X, y)
X_os = pd.DataFrame(data = X_os, columns = column)
y_os = pd.DataFrame(data = y_os, columns = ['Transported'])

print(X_os.shape, y_os.shape)
y_os.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_os, y_os)

test_pred = model.predict(test)
```

```python
df=pd.DataFrame({'Transported':test_pred},index = PassengerId)
```
